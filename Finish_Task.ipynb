```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a DataFrame with random data for demonstration
df = pd.DataFrame(np.random.randn(30, 4).cumsum(0), columns=['Hardness', 'Sulfate', 'Conductivity', 'Trihalomethanes'], index=np.arange(0, 300, 10))

# Plot the DataFrame
df.plot()
plt.title('Random Data Cumulative Sum Plot')
plt.xlabel('Index')
plt.ylabel('Cumulative Sum')
plt.show()

```
